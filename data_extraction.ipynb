```python
# access through openneuro
from cloudpathlib import S3Path, S3Client
from pathlib import Path
from utils import ls

# Set up our cache path:
cache_path = Path('/tmp/cache')
if not cache_path.exists():
    cache_path.mkdir()

# Create the root S3Path for the NYU Retinotopy Dataset:
# here loop through your sub id:
with open('sub.txt') as f:
    sublist = f.readlines()
    
list_all = []
tasks = ['bart', 'bht', 'scap','stopsignal']

for sub in sublist:
    s3_path = 's3://openneuro/ds000030/ds000030_R1.0.5/uncompressed/derivatives/fmriprep/' + sub.strip() + '/func'
    nyu_base_path = S3Path(
        s3_path,
        client=S3Client(
            no_sign_request=True,
            local_cache_dir=cache_path))

    all_files = ls(nyu_base_path)

    for file in all_files:
        for task in tasks:
            if str(file).endswith('-' + task + '_bold_space-MNI152NLin2009cAsym_preproc.nii.gz'):
                #print(file)
                list_all.append(str(file))
            else:
                if str(file).endswith('-' + task + '_bold_confounds.tsv'):
                  #  print(file)
                    list_all.append(str(file))
    
```
